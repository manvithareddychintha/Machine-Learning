```python
#correletion to find the correletaion the columns have to in nuumeric form either encode the columns with numbers or select only numeric columns
```

```python
data_numeric = data.select_dtypes(include=[np.number])  # select numeric columns only
corr_matrix = data_numeric.corr()
print(corr_matrix)
```

```python
#heatmap
```

```python
sns.heatmap(corr_matrix,annot=True)
plt.rcParams['figure.figsize']=(20,7)
plt.show()
```
